# Company's Ideal Customer

## Expolring data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/proj35/marketing_campaign.csv", delimiter="\t")
df.head()
```

```python
df.head().T
```

## Cleaning data

```python
df.isna().sum()
```

```python
df["Income"].fillna(np.mean(df['Income']), inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
df.drop(["ID", "Dt_Customer"], inplace=True, axis=1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Education"] = le.fit_transform(df["Education"])
le.classes_
```

```python
df["Marital_Status"] = le.fit_transform(df["Marital_Status"])
le.classes_
```

```python
df.info()
```

## Preprocessing the model

```python
X = df.drop("Response", axis=1)
y = df["Response"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training a model

```python
from sklearn.ensemble import RandomForestClassifier

rclf = RandomForestClassifier()

rclf.fit(X_train, y_train)
rclf.score(X_test, y_test)
```

```python
from xgboost import XGBClassifier

xgclf = XGBClassifier()

xgclf.fit(X_train, y_train)
xgclf.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knclf = KNeighborsClassifier()

knclf.fit(X_train, y_train)
knclf.score(X_test, y_test)
```
