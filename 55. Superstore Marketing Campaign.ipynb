## Pre-Processing Dataset

```python
df.head().T
```

```python
df.drop("Id", axis=1, inplace=True)

from sklearn.preprocessing import LabelEncoder

df.drop("Dt_Customer", axis=1, inplace=True)

for column in df.keys():
    if pd.api.types.is_object_dtype(df[column]):
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        print("-"*20)
        print(column)
        print(le.classes_)
        print("-"*20)
    else:
        pass

X = df.drop("Complain", axis=1)
y = df["Complain"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
rfclf = RandomForestClassifier()
rfclf.fit(X_train, y_train)
rfclf.score(X_test, y_test)
```

```python
kclf = KNeighborsClassifier().fit(X_train, y_train)
kclf.score(X_test, y_test)
```

```python
svclf = SVC().fit(X_train, y_train)
svclf.score(X_test, y_test)
```

```python
xgbclf = XGBClassifier()
xgbclf.fit(X_train.values, y_train.values)
xgbclf.score(X_test, y_test)
```
