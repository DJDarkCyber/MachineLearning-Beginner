# Motor Bikes in Indian Market

## Getting the data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/proj51/bikes.csv")
df.head()
```

## Cleaning data

```python
df.isna().sum()
```

```python
df.drop("Unnamed: 0", inplace=True, axis=1)
```

```python
import re

pattern = re.compile(r"[₹,]")

df["Price"].replace(pattern, "", inplace=True)
df["Price"]
```

```python
pattern = re.compile(r"[C]")
pattern
```

```python
df["CC"].replace(pattern, "", inplace=True)
df["CC"]
```

```python
pattern = re.compile(r"[Km/l]")
df["Mileage"].replace(pattern, "", inplace=True)
df["Mileage"]
```

```python
pattern = re.compile(r"Kg")
df["Weight"].replace(pattern, "", inplace=True)
df["Weight"]
```

```python
df.head()
```

```python
df["Mileage"].replace(re.compile(r"[Fu Charge]"), "", inplace=True)
```

```python
df["Mileage"] = df["Mileage"].astype("str")
for i in range(0, len(df["Mileage"])):
    cont = df["Mileage"][i]
    print(type(cont))
    if "-" in cont:
        miles = cont.split("-")
        m1 = float(miles[0])
        m2 = float(miles[1])
        miles[0] = m1
        miles[1] = m2
        mean_mile = np.mean(miles)
        df["Mileage"][i] = mean_mile
```

```python
df.drop("Links", inplace=True, axis=1)
```

```python
df["Price"] = df["Price"].astype("int64")
df["CC"] = df["CC"].astype("float64")
df["Mileage"] = df["Mileage"].astype("float64")
df["Weight"] = df["Weight"].astype("float64")
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Model Name"] = le.fit_transform(df["Model Name"])
le.classes_
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df["CC"].fillna(np.mean(df["CC"]), inplace=True)
df["Mileage"].fillna(np.mean(df["Mileage"]), inplace=True)
```

```python
df.isna().sum()
```

## Pre-Processing Data

```python
X = df.drop("Price", axis=1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training the model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
rfr.score(X_test, y_test)
```

```python
knr = KNeighborsRegressor().fit(X_train, y_train)
knr.score(X_test, y_test)
```

```python
svrr = SVR().fit(X_train, y_train)
svrr.score(X_test, y_test)
```

```python
xgbr = XGBRegressor()
xgbr.fit(X_train.values, y_train.values)
xgbr.score(X_test, y_test)
```
