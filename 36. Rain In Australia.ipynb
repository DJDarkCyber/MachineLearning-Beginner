## Cleaning the data

```python
df.isna().sum()
```

```python
df["MinTemp"].fillna(np.mean(df["MinTemp"]), inplace=True)
df["MaxTemp"].fillna(np.mean(df["MaxTemp"]), inplace=True)
df["Rainfall"].fillna(np.mean(df["Rainfall"]), inplace=True)
df["Evaporation"].fillna(0, inplace=True)
df["Sunshine"].fillna(0, inplace=True)
df["WindGustSpeed"].fillna(np.mean(df["WindGustSpeed"]), inplace=True)
df["WindSpeed9am"].fillna(np.mean(df["WindSpeed9am"]), inplace=True)
df["WindSpeed3pm"].fillna(np.mean(df["WindSpeed3pm"]), inplace=True)
df["Humidity9am"].fillna(np.mean(df["Humidity9am"]), inplace=True)
df["Humidity3pm"].fillna(np.mean(df["Humidity3pm"]), inplace=True)
df["Pressure9am"].fillna(np.mean(df["Pressure9am"]), inplace=True)
df["Pressure3pm"].fillna(np.mean(df["Pressure3pm"]), inplace=True)
df["Cloud9am"].fillna(np.mean(df["Cloud9am"]), inplace=True)
df["Cloud3pm"].fillna(np.mean(df["Cloud9am"]), inplace=True)
df["Temp9am"].fillna(np.mean(df["Temp9am"]), inplace=True)
df["Temp3pm"].fillna(np.mean(df["Temp3pm"]), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
for stuff in df.keys()[1:]:
    if pd.api.types.is_object_dtype(df[stuff]) == True:
        le = LabelEncoder()
        df[stuff] = le.fit_transform(df[stuff])
        print("-"*20)
        print(le.classes_)
        print("-"*20)
    else:
        pass
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df["RainToday"] =  df["RainToday"].replace(2, 0)
df["RainToday"].unique()
```

```python
df["RainTomorrow"].replace(2, 0, inplace=True)
df["RainTomorrow"].unique()
```

```python
df["WindGustDir"].replace(16, 13, inplace=True)
df["WindDir9am"].replace(16, 3, inplace=True)
df["WindDir3pm"].replace(16, 9, inplace=True)
```

```python
df["Date"][0].split("-")

years = []
months = []
days = []

for i in range(0, len(df["Date"])):
    split_date = df["Date"][0].split("-")
    years.append(split_date[0])
    months.append(split_date[1])
    days.append(split_date[2])

df["Year"] = years
df["Month"] = months
df["Day"] = days
```

```python
df.drop("Date", axis=1, inplace=True)
df.head()
```

```python
X = df.drop("RainTomorrow", axis=1)
y = df["RainTomorrow"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training the model

```python
from sklearn.ensemble import RandomForestClassifier

rclf = RandomForestClassifier()
rclf.fit(X_train, y_train)
```

```python
rclf.score(X_test, y_test)
```
