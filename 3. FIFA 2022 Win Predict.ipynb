```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("data/proj3/international_matches.csv")
df.head()
```

```python
df.T
```

```python
df.info()
```

```python
df_tmp = df.copy()
```

```python
df_tmp["day"] = pd.to_datetime(df_tmp["date"]).dt.day
df_tmp["month"] = pd.to_datetime(df_tmp["date"]).dt.month
df_tmp["year"] = pd.to_datetime(df_tmp["date"]).dt.year
```

```python
df_tmp.drop(["home_team_goalkeeper_score", "away_team_goalkeeper_score", "home_team_mean_defense_score", "home_team_mean_offense_score", "home_team_mean_midfield_score", "away_team_mean_defense_score", "away_team_mean_offense_score", "away_team_mean_midfield_score", "date"], inplace=True, axis=1)
```

```python
df_tmp.T
```

```python
for stuff in df_tmp.keys():
    if pd.api.types.is_object_dtype(df_tmp[stuff]) == True:
        new_cat = pd.Categorical(df_tmp[stuff])
        df_tmp[stuff + "_cat"] = new_cat.codes
```

```python
df_tmp.keys()
```

```python
df_tmp["neutral_location_cat"] = pd.Categorical(df["neutral_location"]).codes
```

```python
df_tmp.T
```

```python
from sklearn.model_selection import train_test_split

X = df_tmp.drop(["home_team", "away_team", "home_team_continent", "away_team_continent", "city", "country", "shoot_out", "home_team_result", "home_team_result_cat", "tournament", "neutral_location"], axis=1)
y = df_tmp["home_team_result_cat"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
model_rr = RandomForestRegressor()
model_rr.fit(X_train, y_train)
```

```python
model_rr.score(X_test, y_test)
```
