```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D 
from tensorflow.keras.layers import MaxPooling2D 
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
training_gen = ImageDataGenerator(
    zoom_range=0.2,
    shear_range=0.2,
    rescale=1./255,
    horizontal_flip=True,
    validation_split=0.2,
)

training_set = training_gen.flow_from_directory(
    "trainingSet/trainingSet/",
    target_size=(28, 28),
    class_mode='binary',
    batch_size=32,
    subset='training'
)

validation_set = training_gen.flow_from_directory(
    'trainingSet/trainingSet/',
    target_size=(28, 28),
    class_mode='binary',
    batch_size=32,
    subset='validation'
)
```

```python
model = Sequential()

# add convolutional layer with 32 filters, a 3x3 kernel, and ReLU activation
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(28, 28, 3)))

# add a max pooling layer with a 2x2 pool size
model.add(MaxPooling2D(pool_size=(2, 2)))

# add another convolutional layer with 64 filters and a 3x3 kernel
model.add(Conv2D(64, (3, 3), activation='relu'))

# add another max pooling layer with a 2x2 pool size
model.add(MaxPooling2D(pool_size=(2, 2)))

# add a dropout layer with a rate of 0.25
model.add(Dropout(0.25))

# flatten the previous layer output into a 1D tensor
model.add(Flatten())

# add a fully-connected layer with 128 units and ReLU activation
model.add(Dense(128, activation='relu'))

# add another dropout layer with a rate of 0.5
model.add(Dropout(0.5))

# add a final fully-connected layer with 10 units and a softmax activation
# for outputting predictions for the 10 classes
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x=training_set, epochs=25, batch_size=32, validation_data=validation_set)
```
