```python
import pandas as pd
import plotly.express as px
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data/proj12/housing.csv")
df.head()
```

```python
df.isna().sum()
```

```python
fig = px.scatter_3d(df, x="LSTAT", y="MEDV", z="PTRATIO", color_continuous_scale = ["#FF0000", "#0000FF", "#FFFF00", 
                                                  "#00EFFF", "#FF00FF", "#FFFFFF"])
fig.update_traces(marker = dict(size = 3, line = dict(width = 1, color = "#FFFFFF")))
fig.update_layout(template = "plotly_dark", font = dict(family = "PT Sans", size = 12, color = "#97FFFF"))
fig.show()
```

```python
_, ax = plt.subplots(3, figsize=(15, 6))
_.suptitle("Boston")
ax[0].scatter(df["RM"], df["MEDV"])
ax[1].scatter(df["LSTAT"], df["MEDV"])
ax[2].scatter(df["PTRATIO"], df["MEDV"])
```

```python
X = df.drop("MEDV", axis=1)
y = df["MEDV"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
