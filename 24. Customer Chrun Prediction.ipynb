# Customer Chrun Prediction

## Getting the dataset

```python
import pandas as pd
df = pd.read_csv("data/proj24/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.head().T
```

## Cleaning the data

```python
df.isna().sum()
```

```python
import numpy as np
df["TotalCharges"].replace(' ', np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].astype("float64")
```

```python
df.fillna(np.mean(df["TotalCharges"]), inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop("customerID", inplace=True, axis=1)
```

## Preprocessing the data

```python
df.info()
```

```python
df.keys()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for stuff in df.keys():
    if pd.api.types.is_object_dtype(df[stuff]) == True:
        df[stuff] = le.fit_transform(df[stuff])
        print(stuff, " : ")
        print(le.classes_)
        print("-"*20)
        
```

```python
df.head().T
```

## Preparing the data to train in model

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
len(X_train), len(y_train), len(X_test), len(y_test)
```

## Training the model

```python
from sklearn.ensemble import RandomForestClassifier

rclf = RandomForestClassifier()
rclf.fit(X_train, y_train)
rclf.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knclf = KNeighborsClassifier()
knclf.fit(X_train, y_train)
knclf.score(X_test, y_test)
```

```python
from xgboost import XGBClassifier

xgclf = XGBClassifier()
xgclf.fit(X_train, y_train)
```

```python
xgclf.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
svcclf = SVC()
svcclf.fit(X_train, y_train)
svcclf.score(X_test, y_test)
```

RandomForestClassifier ensemble gives 79% accuracy which is better XGBoost also better
